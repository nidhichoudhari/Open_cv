# Edge detection techniques (Sobel and Canny)

```python
# Import dependencies
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the original image
image_color = cv2.imread('test.jpg')  # Keep the original colored version
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)  # Grayscale version for edge detection

# Apply edge detection
gradients_sobelx = cv2.Sobel(image_gray, -1, 1, 0)
gradients_sobely = cv2.Sobel(image_gray, -1, 0, 1)
gradients_sobelxy = cv2.addWeighted(gradients_sobelx, 0.5, gradients_sobely, 0.5, 0)
gradients_laplacian = cv2.Laplacian(image_gray, -1)
canny_output = cv2.Canny(image_gray, 80, 150)

# Show the original image
print("Original Image")
cv2_imshow(image_color)

# Show edge detection results
print("Sobel X")
cv2_imshow(gradients_sobelx)

print("Sobel Y")
cv2_imshow(gradients_sobely)

print("Sobel X + Y")
cv2_imshow(gradients_sobelxy)

print("Canny")
cv2_imshow(canny_output)
```
